```python
def findDate(rule,line):
    invoiceDate = ""
    isDate = False
    dateRegex = ['(([0-9])|([0-2][0-9])|([3][0-1]))\-(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\-\d{4}$',
                '(([0-9])|([0-2][0-9])|([3][0-1]))\-(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\-\d{2}$',
                '(([0-9])|([0-2][0-9])|([3][0-1]))\-(JAN|FEB|MAR|APR|MAY|JUN|JUL|AUG|SEP|OCT|NOV|DEC)\-\d{2}$',
                 '(([0-9])|([0-2][0-9])|([3][0-1]))\-(JAN|FEB|MAR|APR|MAY|JUN|JUL|AUG|SEP|OCT|NOV|DEC)\-\d{4}$',
                '\d{1,2}\/\d{1,2}\/\d{4}$',
                '\d{1,2}\-\d{1,2}\-\d{4}$',
                '\d{1,2}\/\d{1,2}\/\d{2}$',
                '\d{1,2}\-\d{1,2}\-\d{2}$',
                '\d{4}\.\d{1,2}\.\d{1,2}$',
                '\d{4}\-\d{1,2}\-\d{1,2}$',
                '\d{1,2}\.\d{1,2}\.\d{2}$',
                '\d{1,2}\.\d{1,2}\.\d{4}$',
                '\d{1,2}\.\d{1,2}\.\d{4}$',
                '\d{1}\.\d{1,2}\.\d{4}$',
                '\d{1}\-\d{1,2}\-\d{4}$',
                '\d{1}\.\d{1}\.\d{4}$',
                '\d{1}\-\d{1}\-\d{4}$']
    
    #Rule 1.0 -> search date directly into the words
    for word in line.split(' '):
        for dateRe in dateRegex:
            date = re.search(dateRe,word)
            if not date:
                pass
            else:
                invoiceDate = date.group(0)
                isDate = True
                break
        if(isDate == True):
            break
            
    #Rule 1.1 -> rearch date in the whole line by removing all the white space
    if(isDate == False):
        for dateRe in dateRegex:
            line = re.sub(r"\s+", "", line)
            date = re.search(dateRe,line)
            if not date:
                pass
            else:
                invoiceDate = date.group(0)
                isDate = True
                break
            if(isDate == True):
                break
            
    #Rule 1.2 -> search for the date with no delemeter seperation
    if(isDate == False):
        for word in line.split(' '):
            if(word.isdigit() == True and len(word) == 8):
                date = word
                if not date:
                    pass
                else:
                    invoiceDate = date
                    isDate = True
                    break
            if(isDate == True):
                break
                
    return invoiceDate
def extractInvoiceDate(responseJSON):
    lengthOfOutPutJSON=len(responseJSON)
#     lengthOfOutPutJSON = 1000
    dataArr = []
    value = 0
    for i in range (lengthOfOutPutJSON):
        line = responseJSON[i]
        dataArr.append(line)
        
    invoiceDate = " "
    indexOfInvoiceMatching = 0
    invoiceDateFound = False
    logging = False
    isReject = False
    invoiceDateMatchingArr=['INVOICE DATE','Invoice Date','Invoice date','Invoice Date.','InvoiceDate','nvoice Date',
                        'Invoice Date:','Invoice Date','Invoice date:','Dated',
                        'Bill Date','Bill date',
                        'DATE','Invoice No. / Date',
                        'Date:','date:','Date :','date :',
                        'Dt:','Dt :']
    rejectWords = ['PO . Date:','PO .Date:','PO. Date:','PO.Date:']
    for line in dataArr:
        if logging :print(line)
        for dateMatch in invoiceDateMatchingArr:
            isReject = False
            for reject in rejectWords:
                if(line.find(reject)!=-1):
                    if logging :print("reject line :::",line)
                    isReject=True
                    break
            if(line.find(dateMatch)!=-1 and invoiceDateFound == False and isReject == False):
                if logging :print(line,' ',dateMatch)
                indexOfInvoiceMatching = dataArr.index(line)
                
                #Rule One -> find date in the same line 
                invoiceDate = findDate("One",line)
                if(invoiceDate!=""):
                    invoiceDateFound = True
                    break
                #Rule Two -> to find in the next Five lines
                if(invoiceDateFound == False):
                    counterStart = indexOfInvoiceMatching+1
                    counterEnd = indexOfInvoiceMatching + 6
                    if logging :print("printed by index: ",dataArr[indexOfInvoiceMatching])
                    while(counterStart < counterEnd and invoiceDateFound == False): 
                        newLine = dataArr[counterStart]
                        invoiceDate = findDate("Two",newLine)
                        if logging :print(newLine,'_____________newLine')
                        if(invoiceDate!="" and invoiceDate!=" "):
                            invoiceDateFound = True
                            break
                        counterStart = counterStart + 1
        if(invoiceDateFound == True):
            break
#     if(invoiceDateFound == False):
#             print(invoiceDate, "****************\n\n")
    return invoiceDate
```

```python
def extractIFSCCode(responseJSON):
    lengthOfOutPutJSON=len(responseJSON)
    dataArr = []
    value = 0
    for i in range (lengthOfOutPutJSON):
        line = responseJSON[i]
        dataArr.append(line)
        
    IFSCCode = ''
    IFSC = ''
    isIFSC = False
    
    #Rule one -> search normaly without manuplating the output of azure
    for line in dataArr:
#         print(line)
#         line = re.sub('[^A-Za-z0-9.-\\\/]+', '', line)
        line = line.replace(' ','')
        IFSCCode =  re.search("[A-Za-z]{4}\d{7}",line)
        if not IFSCCode:
            pass
        else:
            tempIFSC = re.sub('[^A-Za-z]+', '', IFSCCode.group(0))
            if(tempIFSC.isupper()):
                IFSC = IFSCCode.group(0)
                isIFSC = True
                break
        
    #Rule two -> in rule two the O of azure output is been modified to 0
    if(isIFSC == False):
#         print("Rule Two Started *^*^*^*^*^*^*^*")
        for line in dataArr:
#             print(line)
            line = line.replace(' ','')
            line = line.replace('O','0')
            IFSCCode =  re.search("[A-Za-z]{4}\d{7}",line)
            if not IFSCCode:
                pass
            else:
                tempIFSC = re.sub('[^A-Za-z]+', '', IFSCCode.group(0))
                if(tempIFSC.isupper()):
                    IFSC = IFSCCode.group(0)
                    isIFSC = True
                    break
#     print(IFSC,'**********************************\n\n')
    return IFSC
```

```python
def extractCurrency(responseJSON):
    lengthOfOutPutJSON=len(responseJSON)
    dataArr = []
    value = 0
    for i in range (lengthOfOutPutJSON):
        line = responseJSON[i]
        dataArr.append(line)
    
    currencyMatch = {'INR':['INR', 'Rupees','Rs.'],
                    '$':['$'],
                    '£':['£']}
    
    isCurrency = False
    currency = "INR"
    currencyArray = []
    for line in dataArr:
        for key in currencyMatch:
            valueList = currencyMatch[key]
            for value in valueList:
                if(line.find(value)!=-1):
                    currencyArray.append(key)
    for curr in currencyArray:
        if(curr.find("INR")!=-1):
            currency = "INR"
            isCurrency = True
            break
    if(isCurrency == False):
        for curr in currencyArray:
            if(curr.find("£")!=-1):
                currency = "£"
                isCurrency = True
                break
    if(isCurrency == False):
        for curr in currencyArray:
            if(curr.find("$")!=-1):
                currency = "$"
                isCurrency = True
                break
    return currency
```

```python
def findRateNameBases(dataArr):
    returnRate = {"CGST":0,"SGST":0,"IGST":0}
    CGST = 0
    SGST = 0
    IGST = 0
    CGSTFound = False
    SGSTFound = False
    IGSTFound = False
    foundRate = 0
    for line in dataArr:
#         print(line)
        if(line.find("CGST")!=-1 and CGSTFound == False):
            foundRate=re.sub('[^0-9]+', '', line)
#             print("*************CGST",foundRate)
            foundRate=foundRate.strip()
            if(foundRate == '' or len(foundRate)>=4):
                CGST = 0
            else:
                CGST = float(foundRate)
            if(float(CGST)>0):
                CGSTFound = True
        if(line.find("SGST")!=-1 and SGSTFound == False):
            foundRate=re.sub('[^0-9]+', '', line)
#             print("*************SGST",foundRate)
            foundRate=foundRate.strip()
            if(foundRate == '' or len(foundRate)>=4):
                SGST = 0
            else:
                SGST = float(foundRate)
            if(float(SGST)>0):
                SGSTFound = True
        if(line.find("IGST")!=-1 and IGSTFound == False):
            foundRate=re.sub('[^0-9]+', '', line)
#             print("*************IGST",foundRate)
            foundRate=foundRate.strip()
            if(foundRate == '' or len(foundRate)>=4):
                IGST = 0
            else:
                IGST = float(foundRate)
            if(float(IGST)>0):
                IGSTFound = True
    if(float(CGST)>0):
        returnRate['CGST'] = CGST
    if(float(SGST)>0):
        returnRate['SGST'] = SGST
    if(float(CGST)==0 and float(SGST)==0 and float(IGST)>0):
        returnRate['IGST'] = IGST
    if(float(CGST)!=float(SGST)):
        returnRate['CGST'] = 0
        returnRate['SGST'] = 0
#     print('\n\n\n')
    return returnRate
def extractRate(responseJSON):
    lengthOfOutPutJSON=len(responseJSON)
    dataArr=[]
    returnRate = {"CGST":0,"SGST":0,"IGST":0}
    foundRate = ''
    fullRateArray = []
    minimumRate = 0

    for i in range (lengthOfOutPutJSON):
        line=responseJSON[i]
        dataArr.append(line)
    rateMatchingArray = ['2.5%','5%','6%','9%','12%','14%','18%','28%',
                        '2.50%','5.00%','6.00%','9.00%','12.00%','14.00%','18.00%','28.00%',
                        '2.5 %','5 %','6 %','9 %','12 %','14 %','18 %','28 %',
                        '2.50 %','5.00 %','6.00 %','9.00 %','12.00 %','14.00 %','18.00 %','28.00 %']
    returnRate = findRateNameBases(dataArr)
    if( returnRate['CGST'] == 0 and returnRate['SGST'] == 0 and returnRate['IGST'] == 0):
#         print("######################################## Second Rule")
        for line in dataArr:
#             print(line)
            for word in line.split(' '):
                for rate in rateMatchingArray:
                    if(word.find(rate)!=-1):
                        foundRate=re.sub('[^0-9.]+', '', word)
                        foundRate=foundRate.strip()
                        fullRateArray.append(float(foundRate))
#                         print("***********************************found******************************",word)
        if(len(fullRateArray)==0):
            for line in dataArr:
                for rate in rateMatchingArray:
                    if(line.find(rate)!=-1):
                        foundRate=re.sub('[^0-9.]+', '', rate)
#                         print("line found rate before float::::::::::::",foundRate)
                        fullRateArray.append(float(foundRate))
#                         print("***********************************found second******************************",word)
        if(len(fullRateArray)>0):
            minimumRate = min(fullRateArray)
        if(fullRateArray.count(minimumRate)>1):
            returnRate['CGST'] = minimumRate
            returnRate['SGST'] = minimumRate
            returnRate['IGST'] = 0
        else:
            returnRate['CGST'] = 0
            returnRate['SGST'] = 0
            returnRate['IGST'] = minimumRate
#         print('\n\n')
    return returnRate
```

```python
def GetResponseDictionary(Invoice_Number, Invoice_Date, Currency,rate_CGST,rate_SGST,rate_IGST, GSTIN, IFSC, PAN, IBAN,pincode):
    dict_Response = {}
    
    dict_Response['Invoice Number'] = Invoice_Number 
    dict_Response['Invoice Date'] = Invoice_Date 
    dict_Response['Currency'] = Currency 
    dict_Response['CGST'] = rate_CGST
    dict_Response['SGST'] = rate_SGST
    dict_Response['IGST'] = rate_IGST
    dict_Response['IFSC'] = IFSC
    dict_Response['GSTIN'] = GSTIN 
    dict_Response['PAN'] = PAN 
    dict_Response['IBAN'] = IBAN
    dict_Response['Pincode'] = pincode
    

    
    return dict_Response

    
def Get_Regex_Based_Value(document, regex):
    for line in document:
        value = re.findall(regex,line)
        if value:
            return value
    
    return " "
    


# -----------------------------------------  Main Function   -----------------------------------------


def Join_List_Value(list_Values, seprator = ', '):
    value = ''
    if ( (type(list_Values) == type([])) and (len(list_Values) > 0) ): 
        value = seprator.join(list_Values)

    return value



def Extract_Details(document):
    
    invoice_Number = extractInvoiceNum(document)
    
    # Invoice Date
    invoice_Date = extractInvoiceDate(document)

    
    currency=extractCurrency(document)
    
    rate_CGST = extractRate(document)['CGST']
    rate_SGST = extractRate(document)['SGST']
    rate_IGST = extractRate(document)['IGST']
    # ---------------------------------------------------------------   Regex Based Values


    # GSTIN Number
    list_GSTIN_No = Get_Regex_Based_Value(document,"\d{2}[A-Z]{5}\d{4}[A-Z]{1}[A-Z\d]{1}[Z]{1}[A-Z\d]{1}")
#     list_GSTIN_No = [list_GSTIN_No[0][0]+list_GSTIN_No[0][1]]
    gstin_No = Join_List_Value(list_GSTIN_No)
    
        
    # IFSC Code
#     list_IFSC_Code = Get_Regex_Based_Value(document,'^[A-Z]{4}0[A-Z0-9]{6}$')
    ifsc_Code = extractIFSCCode(document)
        
    # PAN Number
    list_PAN_No = Get_Regex_Based_Value(document,"[A-Z]{3}[ABCFGHLJPTF]{1}[A-Z]{1}[0-9]{4}[A-Z]{1}")
    pan_No = Join_List_Value(list_PAN_No)

    list_IBAN_No = Get_Regex_Based_Value(document,"^(?:(?:IT|SM)\d{2}[A-Z]\d{22}|CY\d{2}[A-Z]\d{23}|NL\d{2}[A-Z]{4}\d{10}|LV\d{2}[A-Z]{4}\d{13}|(?:BG|BH|GB|IE)\d{2}[A-Z]{4}\d{14}|GI\d{2}[A-Z]{4}\d{15}|RO\d{2}[A-Z]{4}\d{16}|KW\d{2}[A-Z]{4}\d{22}|MT\d{2}[A-Z]{4}\d{23}|NO\d{13}|(?:DK|FI|GL|FO)\d{16}|MK\d{17}|(?:AT|EE|KZ|LU|XK)\d{18}|(?:BA|HR|LI|CH|CR)\d{19}|(?:GE|DE|LT|ME|RS)\d{20}|IL\d{21}|(?:AD|CZ|ES|MD|SA)\d{22}|PT\d{23}|(?:BE|IS)\d{24}|(?:FR|MR|MC)\d{25}|(?:AL|DO|LB|PL)\d{26}|(?:AZ|HU)\d{27}|(?:GR|MU)\d{28})$")
    iban_No = Join_List_Value(list_IBAN_No)
    
    list_pincode = Get_Regex_Based_Value(document,'[1-9][0-9]{5}')
    pincode =  Join_List_Value(list_pincode)
        
    output = GetResponseDictionary(invoice_Number, invoice_Date,currency,rate_CGST,rate_SGST,rate_IGST, gstin_No, ifsc_Code, pan_No, iban_No,pincode)

    return output
        
```

```python
def findGST(field):
    fields = copy.deepcopy(field)
    GST_values=[]
    for f in fields:
        o={'CGST':f['CGST'],
          'SGST':f['SGST'],
          'IGST':f['IGST']}
        if o['CGST'] == 0:
            if o['IGST']!=0:
                f['CGST']=o['IGST']/2
        if o['SGST'] == 0:
            if o['IGST']!=0:
                f['SGST']=o['IGST']/2
        if o['IGST'] == 0:
            if o['CGST']!=0:
                f['IGST']=o['CGST'] + o['CGST']
            else:
                f['IGST']=o['SGST'] + o['SGST']  
        GST_values.append(f)
    return GST_values
```
